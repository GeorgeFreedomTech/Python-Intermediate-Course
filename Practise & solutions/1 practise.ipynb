````python
# practise I
# 1.
current_funds, monthly_income, mission_duration = float(input("Enter current expedition funds:")), float(input("Enter monthly resource allocation:")), int(input("Enter mission duration (months):"))
total_resources = 0

# 2.
for month in range(mission_duration): 
    total_resources += monthly_income # Add monthly allocation (savings)

# 3.
# Report results
print(f"Final expedition funds after {mission_duration} months: {current_funds+total_resources}")
print(f"Total resources accumulated during this period: {total_resources}")

# 4. 
APPRECIATION_RATE = 0.015 # 1.5% resource appreciation per month
total_resources = 0

for month in range(mission_duration):
    monthly_income = float(input(f"Enter monthly resource allocation for month {month + 1}:"))
    total_resources += monthly_income
    total_resources += total_resources * APPRECIATION_RATE
    print(f"Total resources accumulated after month {month + 1}: {total_resources}")


# practise II

operative_performance = [
    ("Pathfinder", [85, 90, 88]),
    ("Spectre", [92, 88, 95]),
    ("Vanguard", [78, 82, 80]),
    ("Seeker", [95, 91, 93]),
]

# 1. Level:
for index, (callsign, scores) in enumerate(operative_performance, start=1):
    print(f"{index}. {callsign}: Scores {scores}")

# --

# 2. Level:
for callsign, scores in operative_performance:
    sorted_scores = sorted(scores) # Create new sorted list
    print(f"{callsign}: Sorted Scores {sorted_scores}")
````
