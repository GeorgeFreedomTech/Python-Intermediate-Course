## practice

**1. (`os` module)**
- You are writing a script that needs to log data to a specific file within a specific folder.
- The script should first check if a directory named `mission_logs` exists in the current location.
    - If the directory and a file named `gamma_sector.log` inside it **do not exist**, the script should first create the `mission_logs` directory, then create the `gamma_sector.log` file and write a first line to it, e.g., `"Log initialized: Sector Gamma reconnaissance."`.
    - If they **do exist**, the script should simply **append** a new line to the existing `gamma_sector.log` file, e.g., `"New log entry added."`.
- With every write operation (both initial creation and appending), your script should also add the following information to the log file on new lines:
    - The current date and time (using `datetime`).
    - A list of all files currently in the `mission_logs` directory (using `os.listdir`).
- **Challenge:**
    - With each write log the contents of the **parent directory** (one level up from where the script is running).

**2. (`sys` module)**
- Create a Python script that can be triggered from the command line to handle emergency protocols.
- The script should read command-line arguments using `sys.argv`.
- If the specific string `"--code-red"` is found anywhere among the arguments:
    - The script should print a warning message like `"EMERGENCY PROTOCOL ACTIVATED!"`.
    - It should then create (or overwrite) a file named `emergency_log.txt`.
    - In this file, it should write the alert message and the current `datetime` of the event.
- If the `"--code-red"` argument is not found, the script can simply print a message like `"Standard operations. No alerts."`.

```python
# 1. (`os` module)
import os
import datetime

LOG_DIRECTORY = "mission_logs"
LOG_FILE_NAME = "gamma_sector.log"
LOG_FILE_PATH = os.path.join(LOG_DIRECTORY, LOG_FILE_NAME)

if not os.path.exists(LOG_DIRECTORY):
    os.mkdir(LOG_DIRECTORY)

timestamp_log = f"TIMESTAMP: {str(datetime.datetime.now())}\n"

if os.path.exists(LOG_FILE_PATH):
    with open(LOG_FILE_PATH, "a", encoding="utf-8") as log_file:
        log_file.write("\nNew log entry added.\n")
        log_file.write(timestamp_log)
        log_file.write(f"FILES_IN_LOG_DIR: {str(os.listdir(LOG_DIRECTORY))}\n")
else:
    with open(LOG_FILE_PATH, "w", encoding="utf-8") as log_file:
        log_file.write("Log initialized: Sector Gamma reconnaissance.\n")
        log_file.write(timestamp_log)
        log_file.write(f"FILES_IN_LOG_DIR: {str(os.listdir(LOG_DIRECTORY))}\n")


# challenge:
# DTTO + parent_dir_log
parent_dir_log = f"PARENT_DIR_CONTENTS: {str(os.listdir('..'))}\n"

if os.path.exists(LOG_FILE_PATH):
    with open(LOG_FILE_PATH, "a", encoding="utf-8") as log_file:
        log_file.write("\nNew log entry added.\n")
        log_file.write(timestamp_log)
        log_file.write(parent_dir_log)
        log_file.write(f"FILES_IN_LOG_DIR: {str(os.listdir(LOG_DIRECTORY))}\n")
else:
    with open(LOG_FILE_PATH, "w", encoding="utf-8") as log_file:
        log_file.write("Log initialized: Sector Gamma reconnaissance.\n")
        log_file.write(timestamp_log)
        log_file.write(parent_dir_log)
        log_file.write(f"FILES_IN_LOG_DIR: {str(os.listdir(LOG_DIRECTORY))}\n")


# ---

# 2. (`sys` module)
import sys
import datetime

# This code is intended to be run from a terminal, for example:
# python your_script_name.py --some-arg code-red code-blue initialize
# python your_script_name.py --standard

if "code-red" in sys.argv:
    alert_message = "EMERGENCY PROTOCOL ACTIVATED!"
    print(alert_message)
    
    with open("emergency_log.txt", "w", encoding="utf-8") as emergency_file:
        timestamp = str(datetime.datetime.now())
        emergency_file.write(f"{alert_message}\n")
        emergency_file.write(f"Event timestamp: {timestamp}\n")
else:
    print("Standard operations. No alerts.")
```
