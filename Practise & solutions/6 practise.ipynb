```python
# 1. 
import time

# a)
for now in range(5):
    print(f"Time check {now + 1}, pre-pause:  {time.asctime()}")
    time.sleep(2)
    print(f"Time check {now + 1}, post-pause: {time.asctime()}\n")

# b) 
start_time = time.time()
total_sum = 0

print("Operation starting...")
for number in range(1, 1_000_001):
    total_sum += number
print("Operation finished.")
end_time = time.time()

duration = end_time - start_time
print(f"Total duration of the operation: {duration} seconds.")


# ---

# 2. 

# a)
import calendar

year = int(input("Enter the year (e.g., 2025): "))
month = int(input("Enter the month number (1-12): "))
days_in_month = calendar.monthrange(year, month)[1]

weekend_days_count = 0

for day in range(1, days_in_month + 1):
    if calendar.weekday(year, month, day) >= 5: # 5 and 6 represent the weekend
        weekend_days_count += 1
        
print(f"In month {month}/{year}, there are {weekend_days_count} weekend days.")


# ---

# 3. 
import datetime

# a) 
current_datetime = datetime.datetime.now()
print(f"a) Current datetime: {current_datetime}")

# b) 
mission_start_date = datetime.datetime(year=1935, month=1, day=1)
print(f"b) Mission start object: {mission_start_date}")

# c)
# for future dates use inputs or set them manually
future_event_date = datetime.datetime(year=2041, month=5, day=14)
weekday_name = future_event_date.strftime("%A") # %A" formats the date to the full weekday name
formatted_date_string = future_event_date.strftime("%Y-%m-%d")
print(f"c) The event on {formatted_date_string} will be a {weekday_name}.")

# d) 
rendezvous_date = datetime.datetime(year=2055, month=9, day=18)
time_until_rendezvous = rendezvous_date - datetime.datetime.now()
print(f"d) Days remaining until final rendezvous: {time_until_rendezvous.days}")
```
